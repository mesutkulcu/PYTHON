```python
str_number = input("enter a number")
for x in list(str_number):
  if ord(x) not in range(48,58):
    print(f"It is an invalid entry. Don't use non-numeric, float, or negative values!")
digit= int(str_number)
total = 0
for i in str_number:
  total += int(i)**len(str_number)
if total == digit:
  print(f"{digit} is an Armstrong number")
elif total != digit:
  print(f"{digit} is not an Armstrong number")
```
